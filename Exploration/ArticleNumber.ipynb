```python
import pandas as pd
import os
from pathlib import Path
import matplotlib.pyplot as plt

read_path = r'C:\Users\Julia\Documents\Uni\Master\1\Data-Literacy\Daten\joined_survey_article_standardized_median.csv'
df = pd.read_csv(read_path, delimiter=';', header=0)
```

```python
df
```

```python
# Plotting standardized_war_articles against date
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y')

# Sort the DataFrame by date (if necessary)
df_sorted = df.sort_values('date')

plt.figure(figsize=(10, 6))  
plt.plot(df_sorted['date'], df_sorted['standardized_war_articles'], marker='o', linestyle='-',
         label='Standardized War Articles', color='blue', alpha=0.7) 
plt.xlabel('Date')
plt.ylabel('Standardized War Articles')
plt.title('Standardized War Articles over Time')
plt.legend()
plt.xticks(rotation=45)  

plt.tight_layout()
plt.show()
```

```python
# Plotting Standardized War Articles Above 0.1 Over Time


# Filtering the data to get values greater than 0.1 in 'standardized_war_articles'
filtered_data = df_sorted[df_sorted['standardized_war_articles'] > 0.1]

plt.figure(figsize=(10, 6))  # Adjust the figure size

plt.plot(filtered_data['date'], filtered_data['standardized_war_articles'], marker='o', linestyle='-',
         label='Values above 0.1', color='blue', alpha=0.7)  

plt.xlabel('Date')
plt.ylabel('Standardized War Articles')
plt.title('Standardized War Articles Above 0.1 Over Time')
plt.legend()
plt.xticks(rotation=45) 

plt.tight_layout()
plt.show()
```

```python
# Finding the maximum values for 'standardized_war_articles'
max_values = df_sorted.nlargest(50, 'standardized_war_articles')
max_values_sorted = max_values.sort_values('date')
```

```python
# plot max values
plt.figure(figsize=(10, 6))  

# Plotting standardized_war_articles against date
plt.plot(max_values_sorted['date'], max_values_sorted['standardized_war_articles'], marker='o', linestyle='-',
         label='Standardized War Articles', color='blue', alpha=0.7) 

plt.xlabel('Date')
plt.ylabel('Standardized War Articles')
plt.title('Standardized War Articles over Time')
plt.legend()
plt.xticks(rotation=45)  
plt.show()
```
